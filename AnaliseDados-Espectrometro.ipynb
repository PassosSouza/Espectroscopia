A partir do experimento da dupla fenda conseguimos ter uma primeira ideia de como são padrões de difração e com isso ver que teremos picos de luz nos lugares em que tivermos ondas construtivas. O que pode ser obtido ao ter ondas em um múltiplo inteiro da outra. Isso seria possível se o comprimento de onda da onda for igual a distância entre elas

$\Delta x = n \lambda$,

assim, se olharmos para figura, conseguimos calcular diferença entre ondas se usarmos o triângulo com lados $d$ e $\Delta x$, assim $\sin(\theta) = \frac{\Delta x}{d} \implies \Delta x = d\sin(\theta) $. Resultando na Lei de Bragg,

$\boxed{\lambda = \frac{d \sin(\theta)}{n}}$,

sendo $d$ espaço entre fendas, $\theta$ o ângulo em que ele raio atinge detector e $n$ um inteiro que mostra grau de variação de distância entre ondas.

Se tivessemos várias fendas, veriamos diversas linhas saindo de cada uma delas. Se definirmos cada fenda como um comprimento de onda, conseguimos ver que ao atingirem uma região, estarão em fase se estiverem em fase. Então, pegando o caso em que n=1, conseguimos pegar um caso de fase em que seria possível ver as linhas

Agora vamos importar dados
levando em conta que calculamos $\lambda = a\sin(\theta)$, podemos calcular a incerteza levando em conta a incerteza do ângulo

$\sigma_\lambda =\pm \sqrt{\left( a\cos(\theta)\sigma_\theta\right)^2 }$


$\boxed{\sigma_\lambda =\pm a\cos(\theta) \sigma_\theta}$

# Codigo para fazer figuras de espectro

```python
def comprimento_onda_para_rgb(wavelength):
    """
    Convert a given wavelength (in nanometers) to an RGB color code.
    """
    if wavelength >= 380 and wavelength < 440:
        red = -(wavelength - 440) / (440 - 380)
        green = 0.0
        blue = 1.0
    elif wavelength >= 440 and wavelength < 490:
        red = 0.0
        green = (wavelength - 440) / (490 - 440)
        blue = 1.0
    elif wavelength >= 490 and wavelength < 510:
        red = 0.0
        green = 1.0
        blue = -(wavelength - 510) / (510 - 490)
    elif wavelength >= 510 and wavelength < 580:
        red = (wavelength - 510) / (580 - 510)
        green = 1.0
        blue = 0.0
    elif wavelength >= 580 and wavelength < 645:
        red = 1.0
        green = -(wavelength - 645) / (645 - 580)
        blue = 0.0
    elif wavelength >= 645 and wavelength <= 780:
        red = 1.0
        green = 0.0
        blue = 0.0
    else:
        red = 0.0
        green = 0.0
        blue = 0.0

    return int(red * 255), int(green * 255), int(blue * 255)
```

# HIDROGENIO

```python
a = 10/6*10**(-6)#linhas/m
#o fator n depedene de qual linas estamos vendo, podemos lidar que temos 0 no 0º e para direita, assim como para esquerda o numero vai aumentando

grausi=np.array([14+5/60,15+20/60,17+16/60,23.5+5/60])#hidrogenio

egrausi=np.ones(len(grausi))*4/60
#para contas a seguir defini n como 1, fato de aproximação
lamb = np.ones(len(grausi))
elamb = np.ones(len(grausi))
n=0
for i in range(len(grausi)):
    n=1
    l=abs((a*np.sin(np.radians(grausi[i])))*10**(9))
    if l>=100 and l<= 850:
        lamb[i]=l
        elamb[i]=a*np.cos(np.radians(grausi[i]))*np.radians(egrausi[i])*10**9

#fazer tabela        
print(r'Ângulo (graus) & Comprimento de Onda calculado (nm)\\ ')
for i in range(len(grausi)):
    print(f'{round(grausi[i],3)}({int(round(egrausi[i]*1000,3))}) & {int(round(lamb[i],1))}({int(round(elamb[i],1)*10)}) \\\ ')
print(elamb)
```

Calculando níveis de energia com base nos níveis atômicos do hidrogênio que são, n=1 (13.6 eV) , n=2, (3,4 eV), n=3(1,51 ev), n=4 (0,85 eV) , n=5 (0,54 eV)

```python
NH = np.array([13.6,3.4,1.51,0.85,0.54,0.38])
h = 6.58*10**(-16)*2*np.pi #em eV
c = 3*10**8

lh = np.ones(len(NH)-2)

for i in range(0,len(lh)):
    lh[i] = int(-h*c/(NH[i+2]-NH[1])*10**9)
print('Para ir de um nível de maior energia até o de n=2 teria comprimentos de onda em nm de')
print(lh)
```

Fazer figura do espectro visto

```python
desenho = np.linspace(380,780,401)
Lamb = lamb.copy()
for i in range(len(desenho)):
    u=0
    for j in range(len(Lamb)):
        if int(Lamb[j])==desenho[i]:
            u=-7
            Lamb[j]=100
        else:
            u+=1
    if u>=j:
        desenho[i]=100
        
desenhol = np.linspace(380,780,401)
Lambl = lh.copy()
for i in range(len(desenhol)):
    u=0
    for j in range(len(Lambl)):
        if int(Lambl[j])==desenhol[i]:
            u=-7
            Lambl[j]=100
        else:
            u+=1
    if u>=j:
        desenhol[i]=100

#Exemplo NOVO
fig, ax = plt.subplots(2,1,figsize=(12, 7))
ax[0].set_xlim(0, len(desenho))
ax[0].set_ylim(0, 1)

# Plotar colunas coloridas
for i, comprimento in enumerate(desenho):
    if comprimento!=100:
        red, green, blue = comprimento_onda_para_rgb(comprimento)
        cor_rgb = (red/255, green/255, blue/255)
    else:
        cor_rgb=(0,0,0)
    ax[0].bar(i, 2, color=cor_rgb,width=1)
# Ajustar os limites do eixo x
ax[0].set_xlim(-0.5, len(desenho) - 0.5)
# Configurar os rótulos do eixo x
ax[0].set_xticks(range(len(desenho)))
ax[0].set_xticklabels(desenho)

# Exibir a figura
ax[0].axis('off')
ax[0].set_title(f'Hidrogenio-Experimental', fontsize=25, color='black')

ax[0].set_xlim(0, len(desenho))
ax[0].set_ylim(0, 1)



# Plotar colunas coloridas
for i, comprimento in enumerate(desenhol):
    if comprimento!=100:
        red, green, blue = comprimento_onda_para_rgb(comprimento)
        cor_rgb = (red/255, green/255, blue/255)
    else:
        cor_rgb=(0,0,0)
    ax[1].bar(i, 2, color=cor_rgb,width=1)
# Ajustar os limites do eixo x
ax[1].set_xlim(-0.5, len(desenhol) - 0.5)
# Configurar os rótulos do eixo x
ax[1].set_xticks(range(len(desenhol)))
ax[1].set_xticklabels(desenhol)

# Exibir a figura
ax[1].set_title(f'Hidrogenio-Teorico', fontsize=25, color='black')

ax[1].axis('off')
plt.savefig(f'Experimento-Hidrogenio.png',dpi=100)
```

Assim sabemos o valor de comprimento de onda para cada ângulo. Porém eles fazem sentido? Para responder isso fizemos um programa que possibilita transformar comprimento de onda em um código RGB e assim podemos fazer figura de como seria padrão visto.

```python
lm=np.array([405,408,436,502,546,577,579,615,623])
desenho = np.linspace(380,780,401)
desenhom = np.linspace(380,780,401)
Lamb = lamb.copy()
LambM = lm.copy()
for i in range(len(desenho)):
    u=0
    for j in range(len(Lamb)):
        if abs(int(Lamb[j]))!=int(desenho[i]):
            u+=1
        else:
            Lamb[j]=100
            u=-70
    if u>=j:
        desenho[i]=100
for i in range(len(desenhom)):
    u=0
    for j in range(len(LambM)):
        if abs(int(LambM[j]))!=int(desenhom[i]):
            u+=1
        else:
            LambM[j]=100
            u=-70
    if u>=j:
        desenhom[i]=100

#Exemplo NOVO
fig, ax = plt.subplots(2,1,figsize=(12, 7))
ax[0].set_xlim(0, len(desenho))
ax[0].set_ylim(0, 1)

# Plotar colunas coloridas
for i, comprimento in enumerate(desenho):
    red, green, blue = comprimento_onda_para_rgb(comprimento)
    cor_rgb = (red/255, green/255, blue/255)
    ax[0].bar(i, 2, color=cor_rgb,width=1)
# Ajustar os limites do eixo x
ax[0].set_xlim(-0.5, len(desenho) - 0.5)
# Configurar os rótulos do eixo x
ax[0].set_xticks(range(len(desenho)))
ax[0].set_xticklabels(desenho)

# Exibir a figura
ax[0].set_title(f'Mercurio-Experimental', fontsize=25, color='black')

ax[0].axis('off')

for i, comprimento in enumerate(desenhom):
    if comprimento!=100:
        red, green, blue = comprimento_onda_para_rgb(comprimento)
        cor_rgb = (red/255, green/255, blue/255)
    else:
        cor_rgb=(0,0,0)
    ax[1].bar(i, 2, color=cor_rgb,width=1)
# Ajustar os limites do eixo x
ax[1].set_xlim(-0.5, len(desenhom - 0.5))
# Configurar os rótulos do eixo x
ax[1].set_xticks(range(len(desenhom)))
ax[1].set_xticklabels(desenhom)

# Exibir a figura
ax[1].set_title(f'Mercúrio-Teorico', fontsize=25, color='black')

ax[1].axis('off')
plt.savefig(f'Experimento-Mercurio.png',dpi=100)
```
